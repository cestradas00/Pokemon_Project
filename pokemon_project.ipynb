#### 2.5. We create a new column "Image" where we add the image for each pokemon

```python
from pathlib import Path
image_folder = Path('data/images/')

def get_pokemon_image(row):
    image_path = image_folder / (row['name'].lower() + '.png')
    if image_path.is_file():
        with open(image_path, 'rb') as f:
            return f.read()
    else:
        return None

df['Image'] = df.apply(get_pokemon_image, axis=1)
```

### Statistical tests

#### Difference between the attack for Legendary and Non-Legendary Pokémon

Null Hypothesis (H0): There is no significant difference between the means of the attack stat for Legendary and Non-Legendary Pokémon 

Alternative Hypothesis (Ha): There is a significant difference between the means of the attack stat for Legendary and Non-Legendary Pokémon

H0: μ_legendary_attack = μ_non_legendary_attack                                                                             
Ha: μ_legendary_attack ≠ μ_non_legendary_attack

```python
from scipy.stats import ttest_ind

stat, p_value = ttest_ind(legendary_pokemon['attack'], non_legendary_pokemon['attack'])
print("Attack - t-statistic:", stat, "p-value:", p_value)
```

The t-statistic for the attack base stat is approximately 10.14, and the p-value is very close to zero (5.80e-23). This result indicates that there is a significant difference in the mean attack stat between Legendary and Non-Legendary Pokémon.
We can reject the null hypothesis and conclude that there is strong evidence to suggest that the attack stat differs significantly between Legendary and Non-Legendary Pokémon

#### Difference between the base stats (attack, defense, etc)  for Legendary and Non-Legendary Pokémon

Null Hypothesis (H0): The null hypothesis states that there is no significant difference in the mean base stats (attack, defense, etc.) between Legendary and Non-Legendary Pokémon.

Alternative Hypothesis (Ha): The alternative hypothesis states that there is a significant difference in the mean base stats between Legendary and Non-Legendary Pokémon.

```python
from scipy.stats import f_oneway

stats, p_value = f_oneway(legendary_pokemon['attack'], legendary_pokemon['defense'], non_legendary_pokemon['attack'], non_legendary_pokemon['defense'])
print("ANOVA - F-statistic:", stats, "p-value:", p_value)
```

As the P-Value is extremely low(9.871399314347772e-41),we reject the null hypothesis and conclude that there is a significant difference in the mean base stats between Legendary and Non-Legendary Pokémon

#### Test if the mean HP of Grass-type Pokémon is significantly different from the overall mean HP of all Pokémon

##### One-sample t-test 

Null Hypothesis (H0): The mean HP of Grass-type Pokémon is equal to the overall mean HP of all Pokémon.                      
Alternative Hypothesis (HA): The mean HP of Grass-type Pokémon is not equal to the overall mean HP of all Pokémon. 

H0: μ_grass = μ_all                                                                                        
HA: μ_grass ≠ μ_all

```python
from scipy.stats import ttest_1samp

grass_hp = df_nomega[df_nomega['type_1'] == 'Grass']['hp']

t_stat, p_value = ttest_1samp(grass_hp, df_nomega['hp'].mean())
print(f"t-statistic: {t_stat}, p-value: {p_value}")
```

With a p-value of 0.099, we fail to reject the null hypothesis at a significance level of 0.05. This means that we do not have enough evidence to conclude that the mean HP of Grass-type Pokémon is significantly different from the overall mean HP of all Pokémon.

#### Test if the mean HP of Grass-type Pokémon is significantly different from the mean HP of Fire-type Pokémon.
##### Two-sample t-test



Null Hypothesis (H0): The mean HP of Grass-type Pokémon is equal to the mean HP of Fire-type Pokémon.                         
Alternative Hypothesis (HA): The mean HP of Grass-type Pokémon is not equal to the mean HP of Fire-type Pokémon.

H0: μ_grass = μ_fire                                                                                            
HA: μ_grass ≠ μ_fire

```python
from scipy.stats import ttest_ind

fire_hp = df_nomega[df_nomega['type_1'] == 'Fire']['hp']

t_stat, p_value = ttest_ind(grass_hp, fire_hp)
print(f"t-statistic: {t_stat}, p-value: {p_value}")
```

We fail to reject the null hypothesis. This means that we do not have enough evidence to conclude that there is a significant difference in the mean HP between Grass-type and Fire-type Pokémon.

#### ANOVA test if there is a significant difference in the mean Attack stat across different Pokémon types.

Null Hypothesis (H0): There is no significant difference in the mean Attack stat across different Pokémon types. <br>
Alternative Hypothesis (Ha): There is a significant difference in the mean Attack stat across different Pokémon types.
    
H0: μ_types = μ <br>
HA: μ_types ≠ μ

```python
from scipy.stats import f_oneway

type_attack = {}
for type_name in df_nomega['type_1'].unique():
    type_attack[type_name] = df_nomega[df_nomega['type_1'] == type_name]['attack']

f_stat, p_value = f_oneway(*type_attack.values())
print(f"F-statistic: {f_stat}, p-value: {p_value}")
```

Shows a significant difference in the mean Attack stat across different Pokémon types. Since the p-value (5.21e-15) is much smaller than the typical significance level (α = 0.05), we reject the null hypothesis. We conclude that there is a significant difference in the Attack stat across different Pokémon types.

#### Chi-square test if the distribution of Pokémon types is significantly different between Generation 1 and Generation 2.

Null Hypothesis (H0): There is no significant difference in the distribution of Pokémon types between Generation 1 and Generation 2. <br>
Alternative Hypothesis (Ha): There is a significant difference in the distribution of Pokémon types between Generation 1 and Generation 2.

```python
from scipy.stats import chi2_contingency

contingency_table = pd.crosstab(df_nomega['type_1'], df_nomega['generation'])

chi2_stat, p_value, _, _ = chi2_contingency(contingency_table)
print(f"Chi-square statistic: {chi2_stat}, p-value: {p_value}")
```

Since the p-value (2.28e-06) is much smaller than α = 0.05, we reject the null hypothesis.
We conclude that there is a significant difference in the distribution of Pokémon types between Generation 1 and Generation 2.
This suggests that the proportions of Pokémon types differ significantly between the two generations.

Shows a significant association between Pokémon types and generations. The p-value of 2.28e-06 indicates that there is a relationship between the Pokémon types and the generation they belong to.

#### Chi-square test to analyze the distribution of Legendary Pokémon types compared to non-Legendary Pokémon types

Null Hypothesis (H0): There is no significant difference in the distribution of Pokémon types between Legendary and non-Legendary Pokémon. <br>
Alternative Hypothesis (HA): There is a significant difference in the distribution of Pokémon types between Legendary and non-Legendary Pokémon.

```python
legendary_df = df_nomega[df_nomega['status'].isin(['Mythical', 'Sub Legendary', 'Legendary'])]

contingency_table = pd.crosstab(legendary_df['type_1'], legendary_df['status'])

chi2_stat, p_value, _, _ = chi2_contingency(contingency_table)
print(f"Chi-square statistic: {chi2_stat}, p-value: {p_value}")
```

Since the p-value (0.19999) is greater than α = 0.05, we fail to reject the null hypothesis.
We do not have sufficient evidence to conclude that there is a significant difference in the distribution of Pokémon types between Legendary and non-Legendary Pokémon.
This suggests that the proportions of Pokémon types among Legendary and non-Legendary Pokémon are similar.

There is no significant association between Pokémon type and the status categories "Mythical," "Sub Legendary," and "Legendary" (p-value = 0.200 > 0.05). This suggests that the distribution of Pokémon types among these status categories is not significantly different than what would be expected by random chance.
